**SNAKE WATER GUN**


The Snake-Water-Gun game is a simple game typically played between two players, where each player chooses one of the three items: Snake, Water, or Gun. The game is based on the following rules:

Snake drinks Water, so Snake wins.
Water extinguishes Gun, so Water wins.
Gun kills Snake, so Gun wins.
